# 8. merge 2 dataframe (based on same column)
## # both table ni andar j b same column hoi ani values ne match kre jo math thay to a record merge kri de 

# NOTE:

## focuse on how attribute :
### how : {'left', 'right', 'outer', 'inner', 'cross'}, default 'inner'

   ### * left: use only keys from left frame, similar to a SQL left outer join;
   ### * right: use only keys from right frame, similar to a SQL right outer join;
 ###   * outer: use union of keys from both frames, similar to a SQL full outer join; sort keys lexicographically.
 ###   * inner: use intersection of keys from both frames, similar to a SQL inner join; preserve the order of the left keys.
 ###   * cross: creates the cartesian product from both frames, preserves the order of the left keys.


```python
# example 1 :
df1 = pd.DataFrame({'key1' : [4,5,6,7,8],
              'key2' : [8,9,7,6,8],
              'key3' : [23,54,6,78,3]
})
```

```python
df2 = pd.DataFrame({'key1' : [4,5,1,2,8],
              'key4' : [8,9,47,76,8],
              'key5' : [23,4,46,78,3]
})
```

```python
df1
```

```python
df2
```

```python
# by default inner join 


pd.merge(df1,df2)
```

```python
pd.merge(df1,df2,how='outer')
```

```python
pd.merge(df1,df2, how='left')
```

```python
pd.merge(df1,df2, how='right')
```

```python
pd.merge(df1,df2, how='cross')
```

# 10. apply(user_define or lambda function) :

## whole column pr userdefine function implement krva mate

```python
## task: 
# in df i have Fare in URO...
# my requrinment is...
# (1) make new column named Fare_rupees... 
# (2) convert fare URO in rupees of whole fare URO column...
# (3) store values in Fare_rupees column 

## i can achieve it by using apply function on fare column
```

```python
df
```

```python
df['Fare_rupees'] = df['Fare'].apply(lambda a : a * 80)
```

```python
df
```

```python
# task 2 :

# calculate length of all values of Name column and store it in new column Name_length
df['Name_length'] = df['Name'].apply(lambda a : len(a))
```

```python
df
```

```python
# use external function :
# task 3:
# create new column fare_label store it's value based in fare ...
# if fare<10 then fare_label should be cheap
# if fare>10 and <20 then fare_label should be mid
# if fare>20 then fare_label should be high


def fare_label(x):
    if x<10:
        return 'cheap'
    elif x>10 and x<20:
        return 'mid'
    else:
        return 'high'
```

```python
df['fare_label'] = df['Fare'].apply(fare_label)
```

```python
df
```

```python
# task :
# create new colum age_level based on age
# if age<18 age_level's value should be teen
# if age>18 and <30 age_level's value should be youth
# if age>30 age_level's value should be elder
def age_level(x):
    if x<18:
        return "teen"
    elif x>18 and x<30:
        return "youth"
    else:
        return "elder"
```

```python
df['age_level'] = df['Age'].apply(age_level)
```
